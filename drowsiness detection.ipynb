```python
import torch
import cv2

# Load your trained model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='runs/train/exp10/weights/best.pt', force_reload=True)
model.conf = 0.4  # Confidence threshold

# Start webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLOv5 inference
    results = model(frame)

    # Process results
    for *box, conf, cls in results.xyxy[0]:
        label = f'{model.names[int(cls)]}: {conf:.2f}'
        x1, y1, x2, y2 = map(int, box)

        # Draw bounding box and label
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
        cv2.putText(frame, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

    # Show result
    cv2.imshow("Drowsiness Detection", frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import torch
import cv2
import numpy as np
import time
import winsound  

model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp9/weights/best.pt', force_reload=True)

# Drowsiness detection logic
drowsy_start_time = None
awake_start_time = None
alarm_on = False

def detect_drowsiness(frame, model):
    global drowsy_start_time, awake_start_time, alarm_on

    results = model(frame)
    labels = results.pandas().xyxy[0]['name']

    if 'drowsy' in labels.values:
        if drowsy_start_time is None:
            drowsy_start_time = time.time()
        elif time.time() - drowsy_start_time >= 10 and not alarm_on:
            print("Drowsiness detected for 10+ seconds!")
            winsound.PlaySound("SystemExclamation", winsound.SND_ALIAS | winsound.SND_ASYNC)
            alarm_on = True
        awake_start_time = None
    else:
        if awake_start_time is None:
            awake_start_time = time.time()
        elif time.time() - awake_start_time >= 5 and alarm_on:
            print("Driver awake for 5+ seconds. Stopping alarm.")
            winsound.PlaySound(None, winsound.SND_PURGE)
            alarm_on = False
        drowsy_start_time = None

# Capture from webcam
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    detect_drowsiness(frame, model)
    results = model(frame)
    cv2.imshow('Drowsiness Detection', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import torch
import cv2
import numpy as np
import time
import threading
import winsound


model = torch.hub.load('ultralytics/yolov5', 'custom',path='yolov5/runs/train/exp9/weights/best.pt',force_reload=True)


drowsy_start_time = None
awake_start_time = None
alarm_thread = None
alarm_stop_event = threading.Event()


def play_alarm():
    while not alarm_stop_event.is_set():
        winsound.Beep(1000, 500)
        time.sleep(0.5)

def start_alarm():
    global alarm_thread
    if alarm_thread is None or not alarm_thread.is_alive():
        alarm_stop_event.clear()
        alarm_thread = threading.Thread(target=play_alarm, daemon=True)
        alarm_thread.start()
        print("Alarm started!")

def stop_alarm():
    if alarm_thread and alarm_thread.is_alive():
        alarm_stop_event.set()
        print("Alarm stopped!")

def detect_drowsiness(frame):
    global drowsy_start_time, awake_start_time

    results = model(frame)
    labels = results.pandas().xyxy[0]['name']

    if 'drowsy' in labels.values:
        if drowsy_start_time is None:
            drowsy_start_time = time.time()
        elif time.time() - drowsy_start_time >= 10:
            start_alarm()
        awake_start_time = None
    else:
        if awake_start_time is None:
            awake_start_time = time.time()
        elif time.time() - awake_start_time >= 5:
            stop_alarm()
        drowsy_start_time = None


cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    detect_drowsiness(frame)
    results = model(frame)
    cv2.imshow('Drowsiness Detection', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        stop_alarm()  
        break

cap.release()
cv2.destroyAllWindows()
```
